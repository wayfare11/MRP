```python
import pandas as pd
```

```python
# 读取 Excel 文件的第一个工作表
df_result = pd.read_excel('111(1).xlsx', dtype={'存货编码': str})

# 如果你想读取特定的工作表，可以使用 sheet_name 参数
df_origin = pd.read_excel('7月微波预排计划0630.xlsx', sheet_name='计划排产', dtype={'存货编码': str})
```

```python
print(df_result.columns)
print(df_origin.columns)
```

```python
# 比较 mps 2024年7月
merged_df = pd.merge(df_result, df_origin, on='存货编码', suffixes=('_result', '_origin'))
```

```python
# 合并数据
merged_df = pd.merge(df_result, df_origin, on='存货编码', suffixes=('_result', '_origin'))

# 使用 fillna(0) 处理缺失值
merged_df = merged_df.fillna(0)

result_output = pd.DataFrame()
# 计算 'mps 2024年7月' 和 'MPS\n7月' 列的差值
result_output['存货编码'] = merged_df['存货编码']

result_output['mps 2024年7月_差值'] = merged_df['mps 2024年7月'] - merged_df['MPS\n7月']
result_output['mps 2024年8月_差值'] = merged_df['mps 2024年8月'] - merged_df['MPS\n8月']
result_output['mps 2024年9月_差值'] = merged_df['mps 2024年9月'] - merged_df['MPS\n9月']
result_output['mps 2024年10月_差值'] = merged_df['mps 2024年10月'] - merged_df['MPS\n10月']

result_output['成品库存_差值'] = merged_df['成品库存(现存量查询)'] - merged_df['成品库存 0630']
result_output['解析_差值'] = merged_df['解析库存(现存量查询)'] - merged_df['0630解析']
result_output['在途_差值'] = merged_df['在途(生产订单列表)'] - merged_df['在途 0630']
result_output['月底库存_差值'] = merged_df['月底库存_result'] - merged_df['月底库存_origin']
result_output['7月结余_差值'] = merged_df['2024年7月结余'] - merged_df['7月结余']
result_output['10月结余_差值'] = merged_df['2024年10月结余'] - merged_df['10月结余']

result_output['compare'] = result_output['mps 2024年7月_差值']+result_output['mps 2024年8月_差值']+result_output['mps 2024年9月_差值']+result_output['mps 2024年10月_差值']+result_output['月底库存_差值']-result_output['10月结余_差值']
```
